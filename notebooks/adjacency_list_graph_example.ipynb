# AdjacencyListGraph

Demonstracao de todos os metodos de `AdjacencyListGraph`, um grafo simples **direcionado** representado por **lista de adjacencias**.

| Caracteristica | Detalhe |
|---|---|
| Representacao interna | Lista de `n` dicionarios `{v: peso}` |
| Complexidade de espaco | O(n + m) |
| Lacos `(u, u)` | **Nao permitidos** |
| Arestas multiplas | **Nao suportadas** |
| Indexacao de vertices | Inteiros `[0, num_vertices)` |

---

````python
from graph import AdjacencyListGraph
````

## 3. Criacao do grafo

`AdjacencyListGraph(num_vertices)` inicializa um grafo com `n` vertices e **nenhuma aresta**.  
Internamente, aloca uma lista de `n` dicionarios vazios `{}`.

Os pesos dos vertices sao inicializados como `0.0`.

````python
grafo = AdjacencyListGraph(num_vertices=5)
print("Numero de vertices:", grafo.num_vertices)
print("Pesos dos vertices:", grafo._vertex_weights)
print("Lista de adjacencias:", grafo._adj_list)
````

---
## 4. Informacoes basicas

| Metodo | Complexidade | Descricao |
|---|---|---|
| `get_vertex_count()` | O(1) | Numero de vertices |
| `get_edge_count()` | O(n) | Numero de arestas existentes |
| `is_empty_graph()` | O(n) | `True` se nao ha nenhuma aresta |

````python
print("Vertices:", grafo.get_vertex_count())
print("Arestas:", grafo.get_edge_count())
print("Grafo vazio?", grafo.is_empty_graph())
````

---
## 5. Adicionando e consultando arestas

### `add_edge(u, v)`
Cria a aresta direcionada `u - v` com peso `0.0`.  
E **idempotente**: chamadas repetidas nao criam duplicatas e nao lancam excecao.  
**Complexidade**: O(1) amortizado.

### `has_edge(u, v)`
Retorna `True` se a aresta `u - v` existir.  
**Complexidade**: O(1) amortizado (lookup em dicionario).

Grafo que sera utilizado nos exemplos a seguir:
```
0 - 1
0 - 2
1 - 3
2 - 3
3 - 4
```

````python
grafo.add_edge(0, 1)
grafo.add_edge(0, 2)
grafo.add_edge(1, 3)
grafo.add_edge(2, 3)
grafo.add_edge(3, 4)

print("Arestas apos insercao:", grafo.get_edge_count())
````

````python
print("Lista de adjacencias ({vizinho: peso}):")
for i, neighbors in enumerate(grafo._adj_list):
    print(f"  {i}: {neighbors}")
````

````python
print("0 - 1?", grafo.has_edge(0, 1))   # True
print("1 - 0?", grafo.has_edge(1, 0))   # False - grafo direcionado
print("0 - 3?", grafo.has_edge(0, 3))   # False - aresta nao foi criada
print("3 - 4?", grafo.has_edge(3, 4))   # True
````

### Idempotencia

Chamar `add_edge` para uma aresta ja existente nao duplica a aresta nem altera o peso.

````python
grafo.add_edge(0, 1)  # ja existe
print("Arestas apos add_edge(0, 1) duplicado:", grafo.get_edge_count())  # ainda 5
````

### Lacos sao proibidos

`add_edge(u, u)` lanca `ValueError`.

````python
try:
    grafo.add_edge(2, 2)
except ValueError as e:
    print(f"ValueError: {e}")
````

---
## 6. Removendo arestas

### `remove_edge(u, v)`
Remove a aresta `u - v`, se existir.  
Operacao **silenciosa**: se a aresta nao existir, nenhuma excecao e lancada.  
**Complexidade**: O(1) amortizado.

````python
print("Antes: 2 - 3?", grafo.has_edge(2, 3))

grafo.remove_edge(2, 3)

print("Depois: 2 - 3?", grafo.has_edge(2, 3))
print("Total de arestas:", grafo.get_edge_count())
````

````python
grafo.remove_edge(4, 0)  # aresta inexistente; operacao silenciosa, sem excecao
print("remove_edge em aresta inexistente: nenhuma excecao")

grafo.add_edge(2, 3)  # reinsere para manter o grafo para os proximos exemplos
print("Aresta 2-3 reinserida. Total de arestas:", grafo.get_edge_count())
````

---
## 7. Pesos de vertices

Cada vertice pode carregar um valor numerico (`float`).  
Os pesos sao independentes das arestas e inicializados como `0.0`.

| Metodo | Complexidade | Descricao |
|---|---|---|
| `set_vertex_weight(v, w)` | O(1) | Define o peso do vertice `v` |
| `get_vertex_weight(v)` | O(1) | Retorna o peso do vertice `v` |

````python
grafo.set_vertex_weight(0, 10.5)
grafo.set_vertex_weight(3, 7.0)

for v in range(grafo.num_vertices):
    print(f"  peso do vertice {v}: {grafo.get_vertex_weight(v)}")
````

---
## 8. Pesos de arestas

Arestas sao criadas com peso `0.0` por padrao. `set_edge_weight` permite alterar esse valor.

| Metodo | Complexidade | Descricao |
|---|---|---|
| `set_edge_weight(u, v, w)` | O(1) amortizado | Define o peso da aresta `(u, v)` |
| `get_edge_weight(u, v)` | O(1) amortizado | Retorna o peso da aresta `(u, v)` |

> A aresta deve existir antes de chamar `set_edge_weight`; caso contrario, `ValueError` e lancado.

````python
grafo.set_edge_weight(0, 1, 3.5)
grafo.set_edge_weight(1, 3, 2.0)
grafo.set_edge_weight(3, 4, 8.0)

print("Peso 0 - 1:", grafo.get_edge_weight(0, 1))
print("Peso 1 - 3:", grafo.get_edge_weight(1, 3))
print("Peso 3 - 4:", grafo.get_edge_weight(3, 4))
````

````python
print("Lista de adjacencias com pesos atualizados:")
for i, neighbors in enumerate(grafo._adj_list):
    print(f"  {i}: {neighbors}")
````

---
## 9. Relacoes entre vertices

### `is_successor(u, v)`
`v` e **sucessor** de `u` se existe a aresta `u - v`.

### `is_predecessor(u, v)`
`u` e **predecessor** de `v` se existe a aresta `u - v`.

### `is_incident(u, v, x)`
O vertice `x` e **incidente** a aresta `(u, v)` se `x == u` ou `x == v`.  
Lanca `ValueError` se a aresta nao existir.

> As tres operacoes tem complexidade **O(1) amortizado**.

````python
# is_successor: v e sucessor de u se existe u - v
print("is_successor(0, 1) - 1 e sucessor de 0?", grafo.is_successor(0, 1))  # True
print("is_successor(1, 0) - 0 e sucessor de 1?", grafo.is_successor(1, 0))  # False

print()

# is_predecessor: u e predecessor de v se existe u - v
print("is_predecessor(0, 1) - 0 precede 1?", grafo.is_predecessor(0, 1))
print("is_predecessor(1, 0) - 1 precede 0?", grafo.is_predecessor(1, 0))
````

````python
# is_incident: x e extremo da aresta (u, v)
print("Vertice 0 e incidente a aresta (0-1)?", grafo.is_incident(0, 1, 0))  # True - e origem
print("Vertice 1 e incidente a aresta (0-1)?", grafo.is_incident(0, 1, 1))  # True - e destino
print("Vertice 2 e incidente a aresta (0-1)?", grafo.is_incident(0, 1, 2))  # False

print()

# Aresta inexistente lanca ValueError
try:
    grafo.is_incident(0, 4, 0)  # aresta (0, 4) nao existe
except ValueError as e:
    print(f"ValueError: {e}")
````

## 10. Graus dos vertices

| Metodo | Complexidade | Descricao |
|---|---|---|
| `get_vertex_in_degree(u)` | O(n + m) | No. de arestas que **chegam** a `u` |
| `get_vertex_out_degree(u)` | O(1) | No. de arestas que **partem** de `u` |

> `get_vertex_out_degree` e O(1) porque basta consultar `len(_adj_list[u])`.  
> `get_vertex_in_degree` e O(n + m) porque precisa percorrer todas as listas para contar quem aponta para `u`.

````python
print(f"{'Vertice':<10} {'in_degree':<12} {'out_degree':<12}")
print("-" * 34)
for v in range(grafo.num_vertices):
    in_d = grafo.get_vertex_in_degree(v)
    out_d = grafo.get_vertex_out_degree(v)
    print(f"  {v:<8} {in_d:<12} {out_d:<12}")
````

---
## 11. Relacoes entre arestas

| Conceito | Condicao | Metodo |
|---|---|---|
| **Divergentes** | Mesma **origem**: `u1 == u2` | `is_divergent(u1, v1, u2, v2)` |
| **Convergentes** | Mesmo **destino**: `v1 == v2` | `is_convergent(u1, v1, u2, v2)` |

> Os metodos validam os indices mas **nao** exigem que as arestas existam no grafo.

No grafo de exemplo:
- `(0-1)` e `(0-2)` sao **divergentes** - partem ambas de `0`
- `(1-3)` e `(2-3)` sao **convergentes** - chegam ambas em `3`

````python
# Divergentes - mesma origem
print("(0-1) e (0-2) sao divergentes?", grafo.is_divergent(0, 1, 0, 2))  # True
print("(0-1) e (1-3) sao divergentes?", grafo.is_divergent(0, 1, 1, 3))  # False

print()

# Convergentes - mesmo destino
print("(1-3) e (2-3) sao convergentes?", grafo.is_convergent(1, 3, 2, 3))  # True
print("(0-1) e (0-2) sao convergentes?", grafo.is_convergent(0, 1, 0, 2))  # False
````

## 12. Propriedades globais do grafo

| Metodo | Complexidade | Descricao |
|---|---|---|
| `is_empty_graph()` | O(n) | `True` se nao ha nenhuma aresta |
| `is_connected()` | O(n + m) | `True` se o grafo e **fortemente conectado** |
| `is_complete_graph()` | O(n) | `True` se possui todas as `n*(n-1)` arestas possiveis |

### Conectividade forte

`is_connected()` usa **duas DFS**: uma no grafo original e outra no transposto, ambas a partir do vertice `0`.  
O grafo e fortemente conectado se e somente se as duas DFS visitam todos os vertices.

````python
print("Grafo vazio?", grafo.is_empty_graph())  # False - tem arestas

grafo_sem_arestas = AdjacencyListGraph(num_vertices=3)
print("Grafo sem arestas e vazio?", grafo_sem_arestas.is_empty_graph())  # True
````

````python
# Grafo principal: 0-1, 0-2, 1-3, 2-3, 3-4
# Nao ha caminho de volta para o vertice 0 - nao e fortemente conectado
print("Grafo principal e fortemente conectado?", grafo.is_connected())  # False

# Grafo ciclico: 0-1-2-3-0 - cada vertice alcanca todos os outros
grafo_ciclo = AdjacencyListGraph(num_vertices=4)
grafo_ciclo.add_edge(0, 1)
grafo_ciclo.add_edge(1, 2)
grafo_ciclo.add_edge(2, 3)
grafo_ciclo.add_edge(3, 0)
print("Grafo ciclico 0-1-2-3-0 e fortemente conectado?", grafo_ciclo.is_connected())  # True
````

````python
print("Grafo principal e completo?", grafo.is_complete_graph())  # False
print(f"  Arestas presentes: {grafo.get_edge_count()} / {grafo.num_vertices * (grafo.num_vertices - 1)} possiveis")

print()

# K3: grafo completo com 3 vertices - 3*(3-1) = 6 arestas
n = 3
k3 = AdjacencyListGraph(num_vertices=n)
for u in range(n):
    for v in range(n):
        if u != v:
            k3.add_edge(u, v)

print(f"K{n} (grafo completo) e completo?", k3.is_complete_graph())  # True
print(f"  Arestas: {k3.get_edge_count()} / {n * (n - 1)} possiveis")
````
