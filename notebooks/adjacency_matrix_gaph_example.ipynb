# AdjacencyMatrixGraph

Demonstração de todos os métodos de `AdjacencyMatrixGraph`, um grafo simples **direcionado** representado por **matriz de adjacências**.

| Característica | Detalhe |
|---|---|
| Representação interna | Matriz `n × n` de `float \| None` |
| Complexidade de espaço | O(n²) |
| Laços `(u, u)` | **Não permitidos** |
| Arestas múltiplas | **Não suportadas** |
| Indexação de vértices | Inteiros `[0, num_vertices)` |

## 3. Criação do grafo

`AdjacencyMatrixGraph(num_vertices)` inicializa um grafo com `n` vértices e **nenhuma aresta**.  
Internamente, aloca uma matriz `n × n` preenchida com `None`.

Os pesos dos vértices são inicializados como `0.0`.

````python
grafo = AdjacencyMatrixGraph(num_vertices=5)
print("Número de vértices:", grafo.num_vertices)
print("Pesos dos vértices:", grafo._vertex_weights)
````

---
## 4. Informações básicas

| Método | Complexidade | Descrição |
|---|---|---|
| `get_vertex_count()` | O(1) | Número de vértices |
| `get_edge_count()` | O(n²) | Número de arestas existentes |
| `is_empty_graph()` | O(n²) | `True` se não há nenhuma aresta |

````python
print("Vértices:", grafo.get_vertex_count())
print("Arestas:", grafo.get_edge_count())
print("Grafo vazio?", grafo.is_empty_graph())
````

---
## 5. Adicionando e consultando arestas

### `add_edge(u, v)`
Cria a aresta direcionada `u → v` com peso `0.0`.  
É **idempotente**: chamadas repetidas não criam duplicatas e não lançam exceção.  
**Complexidade**: O(1).

### `has_edge(u, v)`
Retorna `True` se a aresta `u → v` existir.  
**Complexidade**: O(1).

Grafo que será utilizado nos exemplos a seguir:
```
0 → 1
0 → 2
1 → 3
2 → 3
3 → 4
```

````python
grafo.add_edge(0, 1)
grafo.add_edge(0, 2)
grafo.add_edge(1, 3)
grafo.add_edge(2, 3)
grafo.add_edge(3, 4)

print("Arestas após inserção:", grafo.get_edge_count())
````

````python
print("Matriz de adjacências (None = sem aresta, 0.0 = aresta sem peso):")
header = "     " + "  ".join(str(j) for j in range(grafo.num_vertices))
print(header)
for i, row in enumerate(grafo._adj_matrix):
    print(f"  {i}: {row}")
````

````python
print("0 → 1?", grafo.has_edge(0, 1))   # True
print("1 → 0?", grafo.has_edge(1, 0))   # False - grafo direcionado
print("0 → 3?", grafo.has_edge(0, 3))   # False - aresta não foi criada
print("3 → 4?", grafo.has_edge(3, 4))   # True
````

### Idempotência

Chamar `add_edge` para uma aresta já existente não duplica a aresta nem altera o peso.

````python
grafo.add_edge(0, 1)  # já existe
print("Arestas após add_edge(0, 1) duplicado:", grafo.get_edge_count())  # ainda 5
````

### Laços são proibidos

`add_edge(u, u)` lança `ValueError`.

````python
try:
    grafo.add_edge(2, 2)
except ValueError as e:
    print(f"ValueError: {e}")
````

---
## 6. Removendo arestas

### `remove_edge(u, v)`
Remove a aresta `u → v`, se existir.  
Operação **silenciosa**: se a aresta não existir, nenhuma exceção é lançada.  
**Complexidade**: O(1).

````python
print("Antes: 2 → 3?", grafo.has_edge(2, 3))

grafo.remove_edge(2, 3)

print("Depois: 2 → 3?", grafo.has_edge(2, 3))
print("Total de arestas:", grafo.get_edge_count())
````

````python
grafo.remove_edge(4, 0)  # aresta inexistente; operação silenciosa, sem exceção
print("remove_edge em aresta inexistente: nenhuma exceção")

grafo.add_edge(2, 3)  # reinsere para manter o grafo para os próximos exemplos
print("Aresta 2→3 reinserida. Total de arestas:", grafo.get_edge_count())
````

---
## 7. Pesos de vértices

Cada vértice pode carregar um valor numérico (`float`).  
Os pesos são independentes das arestas e inicializados como `0.0`.

| Método | Complexidade | Descrição |
|---|---|---|
| `set_vertex_weight(v, w)` | O(1) | Define o peso do vértice `v` |
| `get_vertex_weight(v)` | O(1) | Retorna o peso do vértice `v` |

````python
grafo.set_vertex_weight(0, 10.5)
grafo.set_vertex_weight(3, 7.0)

for v in range(grafo.num_vertices):
    print(f"  peso do vértice {v}: {grafo.get_vertex_weight(v)}")
````

---
## 8. Pesos de arestas

Arestas são criadas com peso `0.0` por padrão. `set_edge_weight` permite alterar esse valor.

| Método | Complexidade | Descrição |
|---|---|---|
| `set_edge_weight(u, v, w)` | O(1) | Define o peso da aresta `(u, v)` |
| `get_edge_weight(u, v)` | O(1) | Retorna o peso da aresta `(u, v)` |

> A aresta deve existir antes de chamar `set_edge_weight`; caso contrário, `ValueError` é lançado.

````python
grafo.set_edge_weight(0, 1, 3.5)
grafo.set_edge_weight(1, 3, 2.0)
grafo.set_edge_weight(3, 4, 8.0)

print("Peso 0 → 1:", grafo.get_edge_weight(0, 1))
print("Peso 1 → 3:", grafo.get_edge_weight(1, 3))
print("Peso 3 → 4:", grafo.get_edge_weight(3, 4))
````

````python
print("Matriz de adjacências com pesos atualizados:")
for i, row in enumerate(grafo._adj_matrix):
    print(f"  {i}: {row}")
````

## 9. Relações entre vértices

### `is_successor(u, v)`
`v` é **sucessor** de `u` se existe a aresta `u → v`.

### `is_predecessor(u, v)`
`u` é **predecessor** de `v` se existe a aresta `u → v`.

### `is_incident(u, v, x)`
O vértice `x` é **incidente** à aresta `(u, v)` se `x == u` ou `x == v`.  
Lança `ValueError` se a aresta não existir.

> As três operações têm complexidade **O(1)**.

````python
# is_successor: v é sucessor de u se existe u → v
print("is_successor(0, 1) - 1 é sucessor de 0?", grafo.is_successor(0, 1))  # True
print("is_successor(1, 0) - 0 é sucessor de 1?", grafo.is_successor(1, 0))  # False

print()

# is_predecessor: u é predecessor de v se existe u → v
print("is_predecessor(0, 1) - 0 precede 1?", grafo.is_predecessor(0, 1))    # True
print("is_predecessor(1, 0) - 1 precede 0?", grafo.is_predecessor(1, 0))    # False
````

````python
# is_incident: x é extremo da aresta (u, v)
print("Vértice 0 é incidente à aresta (0→1)?", grafo.is_incident(0, 1, 0))  # True - é origem
print("Vértice 1 é incidente à aresta (0→1)?", grafo.is_incident(0, 1, 1))  # True - é destino
print("Vértice 2 é incidente à aresta (0→1)?", grafo.is_incident(0, 1, 2))  # False

print()

# Aresta inexistente lança ValueError
try:
    grafo.is_incident(0, 4, 0)  # aresta (0, 4) não existe
except ValueError as e:
    print(f"ValueError: {e}")
````

## 10. Graus dos vértices

| Método | Complexidade | Descrição |
|---|---|---|
| `get_vertex_in_degree(u)` | O(n) | Nº de arestas que **chegam** a `u` |
| `get_vertex_out_degree(u)` | O(n) | Nº de arestas que **partem** de `u` |

````python
print(f"{'Vértice':<10} {'in_degree':<12} {'out_degree':<12}")
print("-" * 34)
for v in range(grafo.num_vertices):
    in_d = grafo.get_vertex_in_degree(v)
    out_d = grafo.get_vertex_out_degree(v)
    print(f"  {v:<8} {in_d:<12} {out_d:<12}")
````

## 11. Relações entre arestas

| Conceito | Condição | Método |
|---|---|---|
| **Divergentes** | Mesma **origem**: `u1 == u2` | `is_divergent(u1, v1, u2, v2)` |
| **Convergentes** | Mesmo **destino**: `v1 == v2` | `is_convergent(u1, v1, u2, v2)` |

> Os métodos validam os índices mas **não** exigem que as arestas existam no grafo.

No grafo de exemplo:
- `(0→1)` e `(0→2)` são **divergentes** - partem ambas de `0`
- `(1→3)` e `(2→3)` são **convergentes** - chegam ambas em `3`

````python
# Divergentes - mesma origem
print("(0→1) e (0→2) são divergentes?", grafo.is_divergent(0, 1, 0, 2))  # True
print("(0→1) e (1→3) são divergentes?", grafo.is_divergent(0, 1, 1, 3))  # False

print()

# Convergentes - mesmo destino
print("(1→3) e (2→3) são convergentes?", grafo.is_convergent(1, 3, 2, 3))  # True
print("(0→1) e (0→2) são convergentes?", grafo.is_convergent(0, 1, 0, 2))  # False
````

## 12. Propriedades globais do grafo

| Método | Complexidade | Descrição |
|---|---|---|
| `is_empty_graph()` | O(n²) | `True` se não há nenhuma aresta |
| `is_connected()` | O(n²) | `True` se o grafo é **fortemente conectado** |
| `is_complete_graph()` | O(n²) | `True` se possui todas as `n*(n-1)` arestas possíveis |

### Conectividade forte

`is_connected()` usa **duas DFS**: uma no grafo original e outra no transposto, ambas a partir do vértice `0`.  
O grafo é fortemente conectado se e somente se as duas DFS visitam todos os vértices.

````python
print("Grafo vazio?", grafo.is_empty_graph())  # False - tem arestas

grafo_sem_arestas = AdjacencyMatrixGraph(num_vertices=3)
print("Grafo sem arestas é vazio?", grafo_sem_arestas.is_empty_graph())  # True
````

````python
# Grafo principal: 0→1, 0→2, 1→3, 2→3, 3→4
# Não há caminho de volta para o vértice 0 - não é fortemente conectado
print("Grafo principal é fortemente conectado?", grafo.is_connected())  # False

# Grafo cíclico: 0→1→2→3→0 - cada vértice alcança todos os outros
grafo_ciclo = AdjacencyMatrixGraph(num_vertices=4)
grafo_ciclo.add_edge(0, 1)
grafo_ciclo.add_edge(1, 2)
grafo_ciclo.add_edge(2, 3)
grafo_ciclo.add_edge(3, 0)
print("Grafo cíclico 0→1→2→3→0 é fortemente conectado?", grafo_ciclo.is_connected())  # True
````

````python
print("Grafo principal é completo?", grafo.is_complete_graph())  # False
print(f"  Arestas presentes: {grafo.get_edge_count()} / {grafo.num_vertices * (grafo.num_vertices - 1)} possíveis")

print()

# K3: grafo completo com 3 vértices - 3*(3-1) = 6 arestas
n = 3
k3 = AdjacencyMatrixGraph(num_vertices=n)
for u in range(n):
    for v in range(n):
        if u != v:
            k3.add_edge(u, v)

print(f"K{n} (grafo completo) é completo?", k3.is_complete_graph())  # True
print(f"  Arestas: {k3.get_edge_count()} / {n * (n - 1)} possíveis")
````
